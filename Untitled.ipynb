# data cleaning

```python
df.info()
```

### Here We Found 2 Null columns 'stuts'  and 'unnamed1'

```python
#now we drop  this two columns
df.drop(['Status','unnamed1'] , axis= 1, inplace = True)#Here Axis is 1 means Rows and inpalce means all action are run in this sell
```

### Now we Find Null Values

```python
df.isna().sum()
```

#### Now Clean This Null Value

```python
df.shape
```

```python
df.dropna(inplace = True)
```

### Previous Rows are 11251 but now 11239

```python
df.isna().sum()
```

# CHNAGE DATA TYPE

```python
df['Amount'] = df['Amount'].astype('int')
#This Astype is help to convert a int value
```

```python
df['Amount'].dtypes #now this is conver in int datatype
```

```python
#Rename
df.rename(columns ={'Marital_Status' : 'Shadi'}) #this is can not save in next line because i can  not use inplace = 'true'
```

# Now Start  Exploratory Data Analysis (EDA)

## Gender

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender' , data =  df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
```

   #### now this values help to make a plot

```python
seals_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
sns.barplot(x='Gender' , y = 'Amount' , data =seals_gen )
```

summery : From Above Graph we can see that most of the buyers are females and even hte purchasing power of females are greater than Man

# State

```python
df.columns
```

```python
state = sns.countplot(data = df , y ='State')



for  bb in state.containers:
    state.bar_label(bb)
    
```

```python
seals_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
seals_state
```

```python
plt.figure(figsize=(15,4))
sns.barplot(data = seals_state, x= 'State', y = 'Orders')
```

```python
seals_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
plt.figure(figsize=(15,4))
sns.barplot(data = seals_state, x= 'State', y = 'Amount')
```

From above graphs we can see that most of the order & total seals amount are from up and maharasta and karnataka
